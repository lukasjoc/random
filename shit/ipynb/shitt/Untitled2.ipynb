```python
# Padovan Seq

def ped(duration):
    ppp, pp, pc, pn = (1, 1, 1, 1)

    for _ in range(3, duration+1):
        pn = ppp + pp
        ppp = pp       
        pp, pc = pc, pn
        
    return pn
    
    
ped(13)
```

```python
# Padovan Generator

def ped_gen(duration):
    ppp, pp, pc, pn = (1, 1, 1, 1)
    for _ in range(3, duration+1):        
        yield pn
        pn = ppp + pp
        ppp = pp       
        pp, pc = pc, pn
        
ge = ped_gen(13)
next(ge)
```
